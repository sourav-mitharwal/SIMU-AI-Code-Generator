```python
import google.generativeai as genai
import gradio as gr
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
def simu_chat(user_input):

    if not user_input:
        return "Please type a programming request."

    prompt = f"Generate programming code and explain it briefly:\n{user_input}"

    try:
        response = model.generate_content(prompt)
        return response.text

    except Exception as e:
        return str(e)
```

```python
interface = gr.Interface(
    fn=simu_chat,
    inputs="text",
    outputs="text",
    title="SIMU - AI Code Generator",
    description="Ask SIMU to generate programming code."
)
```

```python
interface.launch(share=True)
```
